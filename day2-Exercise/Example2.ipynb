```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X = np.random.randint(0, 100, size=10)#define x value (random num from 0 to 100, only 10 rows)(Independs)
y = np.random.normal() * X + 5 + np.random.randint(0, 10) * X # define y values(depends)
y[6] = y[6] + 10
y[3] = y[3] + -10
y[7] = y[3] + -30
```

```python
plt.scatter(X, y, marker="+") # plot the values (x,y) using scatter plot
```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X.reshape(-1, 1), y.reshape(-1, 1))
print(f"m: {model.coef_}, b: {model.intercept_}") # to find the parameter,weight values that used to defind the model that predict the depends values(y)
```

```python
new_X = np.linspace(X.min(), X.max(), 100) # create 100 points between min and max value of x
y_hat = model.predict(new_X.reshape(-1, 1)) # create model that predect y value (which is depends on x value)
```

```python
plt.scatter(X, y, color="red",marker="+") # Plot the original data points (X , y) 
plt.plot(new_X, y_hat) # Plot the regression line (model predictions)  
```
