```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
df = pd.DataFrame({
    "area": [2600, 3000, 3200, 2600, 4000],
    "bedrooms": [3,4, np.nan, 3, 5],
    "age": [20, 15, 18, 30, 8],
    "price": [550, 565, 610, 595, 760]
})
df
```

```python
# create pipelines that use SimpleImputer method to fill this nan values
pipeline = Pipeline([
    ("fill_nan_values", SimpleImputer())
])
```

```python
# create pipeline that use SimpleImputer method to fill this nan values
# then fits a Linear Regression model
pipeline = Pipeline([
    ("fill_nan_values", SimpleImputer()),
    ("ML_model", LinearRegression())
])
```

```python
pipeline.fit(df.drop("price", axis=1), df["price"]) #fit above pipeline for all coulmns in df and drop "price" column from data frame,axis=1 means drop column not row 
# The "price" column is used as the target variable for training the model.
```

```python
# fit pipelines ,drop price column and use other column as features to predict price column values
pipeline.predict(df.drop("price", axis=1))
```

```python
# Show the statistics (e.g., mean, median, or most frequent values) 
# that the SimpleImputer used to fill missing (NaN) values for each feature
pipeline.named_steps["fill_nan_values"].statistics_
```

```python
pd.Series(
    pipeline.named_steps["fill_nan_values"].statistics_,
    index=df.drop("price", axis=1).columns
)
```
