```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit  # this is the sigmoid function

# Create input values (z) from -10 to +10
z = np.linspace(-10, 10, 300)

# Apply sigmoid (expit) function
sigmoid_output = expit(z)

# Print some sample values
for i in range(0, len(z), 60):
    print(f"z = {z[i]:.2f}, sigmoid(z) = {sigmoid_output[i]:.4f}")

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(z, sigmoid_output, color='red', label='Sigmoid function (expit)')
plt.title("Sigmoid Function from scipy.special.expit")
plt.xlabel("Input z")
plt.ylabel("Sigmoid(z)")
plt.axhline(0.5, color='gray', linestyle='--', label='0.5 threshold')
plt.axvline(0, color='gray', linestyle='--', label='z = 0')
plt.grid(True)
plt.legend()
plt.show()
```
